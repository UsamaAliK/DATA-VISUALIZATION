LINE CHART:
        A line chart is a type of chart that displays information as a series of data points called 'markers' connected by straight line segments. It is often used to visualize trends over time.
       

````python
#DRAWING A LINE PLOT
# Create some imaginary data

apple_yield=[0.895,0.91,0.919,0.926,0.929,0.931]
````


### Explanation of `plt.plot()` and `plt.show()`

- **`plt.plot()`**:  
    This function is used to create a 2D line plot. It takes data points as input and plots them on a graph. By default, it connects the data points with straight lines. You can customize the appearance of the plot by providing additional arguments such as color, line style, markers, etc.

    Example:  
    ```python
    plt.plot(data)
    ```
    This will plot the values in the `data` list on the y-axis, with their indices on the x-axis.

- **`plt.show()`**:  
    This function is used to display the plot. It renders the plot created by `plt.plot()` (or other plotting functions) in the output cell of the Jupyter Notebook. Without calling `plt.show()`, the plot may not be displayed, especially in non-interactive environments.

    Example:  
    ```python
    plt.show()
    ```
    This will display the plot created by the previous plotting commands.


````python
plt.plot(apple_yield)
plt.show()
````


### Adding Labels to Axes in a Plot

Labels are used to describe the axes of a plot, providing context to the data being visualized. They help the viewer understand what the x-axis and y-axis represent.

#### Function to Add Labels:
- **`plt.xlabel()`**:  
    This function is used to set the label for the x-axis.  
    Example:  
    ```python
    plt.xlabel('Year')
    ```

- **`plt.ylabel()`**:  
    This function is used to set the label for the y-axis.  
    Example:  
    ```python
    plt.ylabel('Apple Yield (tons per hectare)')


Adding meaningful labels ensures that the plot is self-explanatory and easier to interpret.

````python
plt.plot(years,apple_yield)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
````

### Chart Title and Legend

#### Chart Title:
The chart title is used to provide a brief description of the plot. It helps the viewer understand the purpose or context of the visualization.  
- **`plt.title()`**:  
    This function is used to set the title of the plot.  
    Example:  
    ```python
    plt.title('Apple and Orange Yield Over the Years')
    ```

#### Legend:
The legend is used to identify different data series in a plot, especially when plotting multiple lines or datasets. It helps distinguish between the plotted elements.  
- **`plt.legend()`**:  
    This function is used to add a legend to the plot. You can specify labels for the data series using the `label` argument in the `plt.plot()` function, and then call `plt.legend()` to display them.  
    Example:  
    ```python
    plt.plot(years, apples, label='Apples')
    plt.plot(years, oranges, label='Oranges')
    plt.legend()
    ```

Adding a title and legend makes the plot more informative and easier to interpret.

````python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title('Apple and Orange Yield Over the Years')
plt.legend(['Apples','Oranges'])
````

### Markers in Matplotlib

Markers are symbols used to highlight individual data points on a plot. They are especially useful when visualizing discrete data points or when you want to emphasize specific points in a line plot.
```python
Example:  
plt.plot(years, apples, marker='o', label='Apples')  
plt.plot(years, oranges, marker='x', label='Oranges')  
plt.xlabel('Year')  
plt.ylabel('Yield (tons per hectare)')  
plt.title('Apple and Orange Yield Over the Years')  
plt.legend()  
plt.show()
```

````python

plt.plot(years,apples,marker='o');
plt.plot(years,oranges,marker='x');
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Apple and Orange Yield Over the Years')
plt.legend(['Apples','Oranges'])
plt.show()
````

### Styling Lines and Markers in Matplotlib

Matplotlib allows you to customize the appearance of lines and markers in your plots. This includes changing the line style, color, width, and marker style.

#### Line Style:
- **`linestyle`**: Specifies the style of the line. Common options include:
    - `'-'`: Solid line (default)
    - `'--'`: Dashed line
    - `'-.'`: Dash-dot line
    - `':'`: Dotted line

#### Line Color:
- **`color`**: Specifies the color of the line. You can use color names (`'red'`, `'blue'`, etc.), hex codes (`'#FF5733'`), or shorthand notations (`'r'`, `'b'`, etc.).

#### Line Width:
- **`linewidth`**: Specifies the thickness of the line. Default is `1.0`.

#### Marker Style:
- **`marker`**: Specifies the style of the marker. Common options include:
    - `'o'`: Circle
    - `'x'`: Cross
    - `'s'`: Square
    - `'d'`: Diamond
    - `'+'`: Plus sign

#### Example:
```python
plt.plot(years, apples, color='blue', linestyle='--', linewidth=2, marker='o', label='Apples')
plt.plot(years, oranges, color='orange', linestyle='-', linewidth=1.5, marker='x', label='Oranges')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title('Apple and Orange Yield Over the Years')
plt.legend()
plt.show()
```

Customizing lines and markers helps make your plots more visually appealing and easier to interpret.
```


````python

plt.plot(years,apples,marker='o',color='red',linestyle='-',linewidth=2,markersize=8,markeredgewidth=2,markeredgecolor='navy');
plt.plot(years,oranges,marker='x',color='orange',linestyle='--',linewidth=2,alpha=1,markersize=8,markeredgewidth=2,markeredgecolor='navy');
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Apple and Orange Yield Over the Years')
plt.legend(['Apples','Oranges'])
plt.show()
````
